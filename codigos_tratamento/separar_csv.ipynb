```python
import pandas as pd

# Carregar o arquivo CSV original
csv_original = pd.read_csv("csvs/tratados/dados_filtrados.csv")
csv_original.head()
```

```python
especialidades = csv_original[["CRM", "Especialidades"]]
especialidades.head()
```

```python
especialidades = especialidades.assign(Especialidades=especialidades["Especialidades"].str.split(",")).explode("Especialidades")
especialidades["Especialidades"] = especialidades["Especialidades"].str.strip()
# especialidades = especialidades.drop(columns=["especialidades"])
especialidades.head()
```

```python
especialidades.to_csv('csvs/final/f_especialidades.csv', index=False)
especialidades.head()
```

```python
medicos = csv_original.drop(columns=["Especialidades"])
medicos.head()
```
