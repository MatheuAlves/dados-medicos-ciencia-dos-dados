```python
import pandas as pd

# Carregar o arquivo CSV original
csv_original = pd.read_csv("csvs/tratados/medicos_mg.csv")
csv_original.head()
```

```python
# Carregar o arquivo de especialidades
especialidades = pd.read_csv("csvs/tratados/especialidades.csv")
especialidades.head()
```

```python
# Garantir que as especialidades de interesse estão em uma lista
especialidades_interesse = especialidades['Especialidade'].str.upper().str.strip().tolist()
print(especialidades_interesse)
```

```python
# Função para remover tudo antes do prefixo
def formatar_especialidades(especialidades_str):
    prefixo = "Especialidades/Áreas de Atuação:,"
    sufixo = ", Visto(s) Provisório(s) ativo(s):"
    
    # Remover o prefixo, se existir
    if prefixo in especialidades_str:
        especialidades_str = especialidades_str.split(prefixo, 1)[1]
    
    # Remover o sufixo, se existir
    if sufixo in especialidades_str:
        especialidades_str = especialidades_str.split(sufixo, 1)[0]
    
    # Retornar a string limpa
    return especialidades_str.strip()


# Aplicar a formatação e filtragem na coluna de especialidades
csv_original['Especialidades Formatadas'] = csv_original['Especialidades'].apply(formatar_especialidades)
csv_original.head()
```

```python
# Filtrar apenas as linhas onde há especialidades de interesse
csv_filtrado = csv_original[csv_original['Especialidades Formatadas'] != ""]
csv_filtrado.head()
```

```python
# Remover a coluna 'nome_coluna'
csv_filtrado = csv_filtrado.drop(columns=['Especialidades'])
# Renomear a coluna 'coluna_antiga' para 'coluna_nova'
csv_filtrado = csv_filtrado.rename(columns={'Especialidades Formatadas': 'Especialidades'})
csv_filtrado = csv_filtrado.rename(columns={'cidade': 'Cidade'})
csv_filtrado['Especialidades'] = csv_filtrado['Especialidades'].replace('Médico sem especialidade registrada.', 'CLÍNICO GERAL')
csv_filtrado.head()
```

```python
import re

def formatar_especialidades(especialidades_str):
    # Remover o padrão ' - RQE Nº: XXXXX'
    especialidades_str = re.sub(r' - RQE Nº: \d+', '', especialidades_str)
    # Remover qualquer texto entre parênteses e o espaço anterior a eles
    especialidades_str = re.sub(r'\s?\(.*?\)', '', especialidades_str)
    # Dividir a string por vírgulas e retornar a lista
    especialidades_lista = [esp.strip() for esp in especialidades_str.split(',')]
    # Transformar a lista em uma string, separando por vírgulas
    return ', '.join(especialidades_lista)

# Aplicar a função na coluna 'Especialidades'
csv_filtrado['Especialidades'] = csv_filtrado['Especialidades'].apply(formatar_especialidades)
csv_filtrado.head()
```

```python

# Salvar o resultado em um novo arquivo CSV
csv_filtrado.to_csv("csvs/tratados/dados_filtrados.csv", index=False)
csv_filtrado.head()
```
